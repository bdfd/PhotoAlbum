```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')

large = 22; med = 16; small = 12
params = {'axes.titlesize': large,
          'legend.fontsize': med,
          'figure.figsize': (16, 10),
          'axes.labelsize': med,
          'axes.titlesize': med,
          'xtick.labelsize': med,
          'ytick.labelsize': med,
          'figure.titlesize': large}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style("white")
%matplotlib inline
```

# Part 1 - Correlation（关联）
Correlation diagrams are used to show relationships between two or more variables. In other means, demonstrate one variable changes with respect to another.

## Scatter plot（散点图）
A scatter plot is a classic and basic diagram used to study the relationship between two variables.

```python
# Import dataset
path1 = 'https://raw.githubusercontent.com/bdfd/Matplotlib_Seaborn_Visualization_World/main/data/midwest_filter.csv'
# sp = scatter plot(you can assign any shortcut)
sp = pd.read_csv(path1)
```

```python
# Create as many colors as there are unique category in dataset midwest filter
categories = np.unique(sp['category'])
colors = [plt.cm.tab10(i/float(len(categories)-1)) for i in range(len(categories))]
```

```python
# Draw Plot for Each Category
plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')

# "c=" 修改为 "cmap="
for i, category in enumerate(categories):
    plt.scatter('area', 'poptotal',
                data=sp.loc[sp.category==category, :],
                s=20, cmap=colors[i], label=str(category))

# Decorations
plt.gca().set(xlim=(0.0, 0.1), ylim=(0, 90000),
              xlabel='Area', ylabel='Population')

plt.xticks(fontsize=12); plt.yticks(fontsize=12)
plt.title("Midwest Area vs Population", fontsize=20)
plt.legend(fontsize=12)    
plt.show()    
```
